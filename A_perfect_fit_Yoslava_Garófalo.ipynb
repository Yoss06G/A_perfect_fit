## 2.2. Descripción de los Datos

Se procede a caracterizar técnicamente el volumen, formato y propiedades de los conjuntos de datos:

* **Estructura Cuantitativa (`train.csv` / `test.csv`):** Identificación del número de observaciones ($N$) y variables. Se verifica que el archivo de entrenamiento contiene variables clave: el identificador único (`ID`) de tipo categórico/nominal y la variable objetivo continua, que representa el porcentaje de calificación o adaptabilidad (escala de 0 a 100).
* **Estructura de Texto (Hojas de Vida y Perfil del Puesto):** Medición de la longitud de los documentos en términos de recuento de caracteres y palabras individuales (*tokens*). Se mapea la distribución formal de las hojas de vida para identificar variables textuales implícitas (secciones de experiencia, educación, y habilidades técnicas).


````python
# ==============================================================================
# 1. ESTRUCTURA CUANTITATIVA (train.csv / test.csv)
# ==============================================================================
print("=" * 70)
print("ESTRUCTURA CUANTITATIVA")
print("=" * 70)

# Diccionario para automatizar el reporte de los DataFrames
datasets = {"train.csv": df_train, "test.csv": df_test}

for nombre, df in datasets.items():
    print(f"\n Conjunto de Datos: {nombre}")
    print(f"  - Número de observaciones (N): {df.shape[0]}")
    print(f"  - Número de variables: {df.shape[1]}")

    # Verificar presencia y tipo de columnas clave
    print("  - Verificación de variables clave:")
    if "ID" in df.columns:
        print(
            f"    * 'CandidateID': Detectado. Tipo de dato: {df['ID'].dtype} (Esperado: categórico/nominal)"
        )
    else:
        print("    * 'CandidateID': No encontrado en las columnas.")

# Debido a que es un conjunto de datos sin la existencia de la variable target presente
# en el conjunto entreamiento (porcentaje de afinidad para el rol) se realizarán modelos
# de aprendizaje no supervisado

# ==============================================================================
# 2. ESTRUCTURA DE TEXTO
# ==============================================================================
print("\n" + "=" * 70)
print("ESTRUCTURA DE TEXTO")
print("=" * 70)


# Función auxiliar para extraer métricas de un archivo PDF
def analizar_pdf(ruta_pdf):
    try:
        reader = PdfReader(ruta_pdf)
        texto_completo = ""
        for pagina in reader.pages:
            texto_pag = pagina.extract_text()
            if texto_pag:
                texto_completo += texto_pag + "\n"

        caracteres = len(texto_completo)
        # Tokenización estándar por espacios y caracteres no alfanuméricos
        palabras = len(re.findall(r"\b\w+\b", texto_completo))
        return caracteres, palabras, texto_completo
    except Exception as e:
        return 0, 0, ""


# A. Análisis del Perfil del Puesto (Job description.pdf)
if os.path.exists(path_pdf):
    caracteres_job, palabras_job, texto_job = analizar_pdf(path_pdf)
    print(f"\n Perfil del Puesto ('{os.path.basename(path_pdf)}'):")
    print(f"  - Longitud en caracteres: {caracteres_job:,}")
    print(f"  - Recuento de palabras (Tokens): {palabras_job:,}")
else:
    print(f"\n No se encontró el archivo PDF en: {path_pdf}")


# B. Análisis Distributivo de las Hojas de Vida (Carpetas)
def analizar_carpeta_resumes(ruta_carpeta, nombre_carpeta):
    if not os.path.exists(ruta_carpeta):
        print(f"\n La carpeta {nombre_carpeta} no existe en la ruta especificada.")
        return

    archivos = [
        f for f in os.listdir(ruta_carpeta) if f.lower().endswith(".pdf")
    ]
    total_archivos = len(archivos)

    print(f"\n Carpeta de Hojas de Vida: '{nombre_carpeta}'")
    print(f"  - Total de documentos PDF encontrados: {total_archivos}")

    if total_archivos > 0:
        metricas = []
        for arch in archivos:
            ruta_completa = os.path.join(ruta_carpeta, arch)
            caracteres, palabras, _ = analizar_pdf(ruta_completa)
            metricas.append({"caracteres": caracteres, "palabras": palabras})

        # Convertir a DataFrame temporal para calcular estadísticas descriptivas
        df_metricas = pd.DataFrame(metricas)

        print("  - Distribución del volumen de texto (Promedios y Dispersión):")
        print(
            f"    * Caracteres por documento ➔ Promedio: {df_metricas['caracteres'].mean():.1f} | Desv. Est.: {df_metricas['caracteres'].std():.1f}"
        )
        print(
            f"    * Palabras (Tokens) por doc  ➔ Promedio: {df_metricas['palabras'].mean():.1f} | Desv. Est.: {df_metricas['palabras'].std():.1f}"
        )
        print(
            f"    * Rango de palabras          ➔ Mínimo: {df_metricas['palabras'].min()} | Máximo: {df_metricas['palabras'].max()}"
        )


# Ejecutar el análisis para las dos carpetas de currículums
analizar_carpeta_resumes(path_train_resumes, "trainResumes")
analizar_carpeta_resumes(path_test_resumes, "testResumes")
````

## 2.3. Exploración de los Datos (EDA)

La exploración combina la teoría estadística descriptiva para la variable continua con técnicas de procesamiento de lenguaje natural (NLP) para el texto:

### Análisis Estadístico de la Variable Objetivo:

1. Se evalúa la distribución
empírica del porcentaje de adaptabilidad en `train.csv`.

2. Se calculan las medidas de tendencia central (media, mediana), de dispersión (varianza, desviación estándar) y de forma (asimetría de Fisher y curtosis).

 Esto permite determinar si la variable sigue una distribución normal o si presenta sesgos que requieran transformaciones matemáticas posteriores.



````python
df_train.info()
````

````python

# Configuración estética de los gráficos
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# ==============================================================================
# ANÁLISIS ESTADÍSTICO DE LA VARIABLE OBJETIVO (Match Percentage)
# ==============================================================================

# Definimos el nombre exacto de la variable objetivo
col_target = "Match Percentage"

# Verificar que la columna exista en el DataFrame de entrenamiento
if col_target in df_train.columns:

    # 1. Extracción de la serie de datos (eliminando nulos si existieran)
    target_data = df_train[col_target].dropna()

    # 2. Cálculo de Medidas Estadísticas Descriptivas
    media = target_data.mean()
    mediana = target_data.median()
    varianza = target_data.var()
    desv_estandar = target_data.std()
    minimo = target_data.min()
    maximo = target_data.max()

    # Medidas de forma (SciPy usa definiciones sesgadas por defecto, ajustamos el cálculo)
    asimetria_fisher = skew(target_data, bias=False)
    curtosis_person = kurtosis(
        target_data, bias=False
    )  # Curtosis en exceso (Normal = 0)

    # Mostrar reporte analítico por consola
    print("=" * 70)
    print(f"CARACTERIZACIÓN ESTADÍSTICA DE: {col_target}")
    print("=" * 70)
    print(f" Medidas de Tendencia Central:")
    print(f"  - Media (Promedio): {media:.4f}%")
    print(f"  - Mediana:          {mediana:.4f}%")
    print(f"\n Medidas de Dispersión:")
    print(f"  - Varianza:         {varianza:.4f}")
    print(f"  - Desviación Est.:  {desv_estandar:.4f}%")
    print(f"  - Rango Dinámico:   [{minimo}% - {maximo}%]")
    print(f"\n Medidas de Forma:")
    print(f"  - Asimetría de Fisher: {asimetria_fisher:.4f}")
    print(
        f"  - Curtosis (Exceso):   {curtosis_person:.4f} (Donde 0 indica Mesocúrtica)"
    )
    print("=" * 70)

    # 3. Interpretación Teórica Automatizada para el Reporte
    print("\n NOTAS PARA LA CARACTERIZACIÓN EN EL REPORTE:")
    if abs(asimetria_fisher) < 0.5:
        print(
            "  - Distribución aproximadamente simétrica con respecto a su media."
        )
    elif asimetria_fisher >= 0.5:
        print(
            "  - Sesgo positivo (asimetría a la derecha): Cola alargada hacia porcentajes altos."
        )
    else:
        print(
            "  - Sesgo negativo (asimetría a la izquierda): Cola alargada hacia porcentajes bajos."
        )

    if abs(curtosis_person) < 0.5:
        print(
            "  - Comportamiento mesocúrtico (proximidad a una distribución normal estándar)."
        )
    elif curtosis_person > 0.5:
        print(
            "  - Comportamiento leptocúrtico: Alta concentración de datos en la región central (colas pesadas)."
        )
    else:
        print(
            "  - Comportamiento platicúrtico: Distribución aplanada con dispersión uniforme de adaptabilidad."
        )

    # 4. Visualización Gráfica Multipanel (Histograma + KDE + Boxplot)
    fig, axes = plt.subplots(
        2, 1, sharex=True, gridspec_kw={"height_ratios": (0.2, 0.8)}
    )

    # Diagrama de Caja (Boxplot) para detección de outliers analíticos
    sns.boxplot(x=target_data, ax=axes[0], color="#5A9")
    axes[0].set(xlabel="")
    axes[0].set_title(
        f"Distribución Empírica de {col_target}", fontsize=14, pad=10
    )

    # Histograma con estimación de densidad por kernel (KDE)
    sns.histplot(target_data, kde=True, ax=axes[1], color="#4C72B0", stat="density")

    # Líneas de referencia para la tendencia central
    axes[1].axvline(
        media,
        color="red",
        linestyle="--",
        linewidth=1.5,
        label=f"Media: {media:.2f}%",
    )
    axes[1].axvline(
        mediana,
        color="green",
        linestyle="-.",
        linewidth=1.5,
        label=f"Mediana: {mediana:.2f}%",
    )

    axes[1].set_xlabel("Porcentaje de Adaptabilidad (%)", fontsize=12)
    axes[1].set_ylabel("Densidad Probabilística", fontsize=12)
    axes[1].legend(loc="upper left")

    plt.tight_layout()
    plt.show()

else:
    print(
        f" Error: No se encontró la columna '{col_target}' en df_train. Verificar los nombres con df_train.columns"
    )
````

A pesar de la falta de normalidad y la presencia de esta bimodalidad, se determina conservar la variable objetivo en su escala empírica original. Aplicar transformaciones matemáticas estabilizadoras (como una transformación logarítmica o Box-Cox) alteraría la escala porcentual lineal ($0-100\%$), destruyendo la interpretabilidad directa del modelo en el contexto del negocio (evaluar directamente el porcentaje real de adaptabilidad del perfil al rol). 2. Se realiza un análisis de frecuencias (*Term Frequency*) sobre el texto limpio y normalizado. Se generan nubes de palabras y gráficos de barras de los $n$-gramas más frecuentes (bi-gramas y tri-gramas) tanto en las hojas de vida como en el archivo `Job description.pdf`.

Esto permite contrastar visual y estadísticamente si los términos de alta relevancia técnica requeridos (ej. *Machine Learning*, *Deep Learning*, *NLP*) emergen de forma natural en los currículos de los candidatos.

````python
# 1. Configuración de Stopwords en Español e Inglés (por si acaso los CVs son bilingües)
stop_words_es = set(stopwords.words("spanish"))
stop_words_en = set(stopwords.words("english"))
# Unimos ambos conjuntos y añadimos variaciones comunes en entornos corporativos
all_stopwords = stop_words_es.union(stop_words_en).union(
    {"correo", "teléfono", "telefono", "experiencia", "año", "años", "perfil"}
)

# 2. Función para extraer, limpiar y normalizar texto de un PDF
def procesar_y_limpiar_pdf(ruta_pdf):
    try:
        reader = PdfReader(ruta_pdf)
        texto_sucio = ""
        for pagina in reader.pages:
            contenido_pag = pagina.extract_text()
            if contenido_pag:
                texto_sucio += contenido_pag + " "

        # Normalización y Limpieza con Expresiones Regulares
        texto = texto_sucio.lower()  # Todo a minúsculas
        texto = re.sub(
            r"https?://\S+|www\.\S+", "", texto
        )  # Eliminar URLs y enlaces
        texto = re.sub(
            r"\S+@\S+", "", texto
        )  # Eliminar correos electrónicos
        texto = re.sub(
            r"[^a-záéíóúüñ\s]", "", texto
        )  # Remover números y caracteres especiales/puntuación
        texto = re.sub(r"\s+", " ", texto).strip()  # Colapsar espacios en blanco

        # Remover Stopwords de forma explícita
        palabras_filtradas = [
            w for w in texto.split() if w not in all_stopwords and len(w) > 2
        ]
        return " ".join(palabras_filtradas)
    except Exception as e:
        # En caso de archivos corruptos o ilegibles, saltar sin romper el bucle
        return ""


# 3. Pipeline de Extracción Masiva para las Carpetas de Currículums
def compilar_texto_carpeta(ruta_carpeta):
    corpus = []
    if os.path.exists(ruta_carpeta):
        archivos = [
            f for f in os.listdir(ruta_carpeta) if f.lower().endswith(".pdf")
        ]
        for arch in archivos:
            ruta_completa = os.path.join(ruta_carpeta, arch)
            texto_limpio = procesar_y_limpiar_pdf(ruta_completa)
            if texto_limpio:
                corpus.append(texto_limpio)
    return corpus


# --- PROCESAMIENTO ---
print("Procesando perfil del puesto (Job Description)...")
texto_job_limpio = (
    procesar_y_limpiar_pdf(path_pdf) if os.path.exists(path_pdf) else ""
)

print("Procesando masivamente currículums de Entrenamiento...")
corpus_train_resumes = compilation_train = compilar_texto_carpeta(
    path_train_resumes
)

print("Procesando masivamente currículums de Prueba...")
corpus_test_resumes = compilar_texto_carpeta(path_test_resumes)

# Combinamos todo el corpus de hojas de vida para tener un análisis global de candidatos
todo_el_corpus_resumes = corpus_train_resumes + corpus_test_resumes
texto_resumes_unificado = " ".join(todo_el_corpus_resumes)

print("Extracción y normalización NLP completada con éxito.")


# 4. Función para graficar N-Gramas (Bi-gramas y Tri-gramas)
def graficar_ngramas(corpus_lista, rango_ngram, titulo, color):
    # Si recibimos un string plano (como el perfil del puesto), lo metemos en una lista
    if isinstance(corpus_lista, str):
        corpus_lista = [corpus_lista]

    if not corpus_lista or len(" ".join(corpus_lista)) == 0:
        print(f"No hay texto suficiente para generar n-gramas en: {titulo}")
        return

    vectorizer = CountVectorizer(ngram_range=rango_ngram, max_features=15)
    transformada = vectorizer.fit_transform(corpus_lista)

    # Sumar las frecuencias de cada término
    frecuencias = transformada.sum(axis=0).A1
    palabras = vectorizer.get_feature_names_out()

    df_frec = (
        pd.DataFrame({"Término": palabras, "Frecuencia": frecuencias})
        .sort_values(by="Frecuencia", ascending=False)
        .reset_index(drop=True)
    )

    # Graficar
    sns.barplot(
        x="Frecuencia", y="Término", data=df_frec, palette=color, hue="Término"
    )
    plt.title(titulo, fontsize=12, fontweight="bold")
    plt.xlabel("Conteo de Frecuencias")
    plt.ylabel("")


# 5. Generación de Visualizaciones Multigraph
plt.figure(figsize=(16, 12))

# --- FILA 1: JOB DESCRIPTION (Perfil del Puesto) ---
plt.subplot(2, 2, 1)
graficar_ngramas(texto_job_limpio, (2, 2), "Bi-gramas: Job Description", "Reds_r")

plt.subplot(2, 2, 2)
graficar_ngramas(
    texto_job_limpio, (3, 3), "Tri-gramas: Job Description", "Oranges_r"
)

# --- FILA 2: HOJAS DE VIDA DE CANDIDATOS (Global) ---
plt.subplot(2, 2, 3)
graficar_ngramas(
    todo_el_corpus_resumes, (2, 2), "Bi-gramas: Hojas de Vida (Todos)", "Blues_r"
)

plt.subplot(2, 2, 4)
graficar_ngramas(
    todo_el_corpus_resumes,
    (3, 3),
    "Tri-gramas: Hojas de Vida (Todos)",
    "Purples_r",
)

plt.tight_layout()
plt.show()


# 6. Generación de Nubes de Palabras (WordClouds)
def mostrar_wordcloud(texto, titulo, mapa_color):
    if len(texto) > 0:
        wc = WordCloud(
            width=800,
            height=400,
            background_color="white",
            colormap=mapa_color,
            max_words=100,
        ).generate(texto)
        plt.figure(figsize=(10, 5))
        plt.imshow(wc, interpolation="bilinear")
        plt.title(titulo, fontsize=14, fontweight="bold", pad=15)
        plt.axis("off")
        plt.show()
    else:
        print(f"⚠️ Texto vacío para la nube de palabras de: {titulo}")


# Desplegar nubes de palabras individuales
mostrar_wordcloud(
    texto_job_limpio, "Nube de Palabras: Requerimientos del Puesto", "autumn"
)
mostrar_wordcloud(
    texto_resumes_unificado, "Nube de Palabras: Corpus de Candidatos", "viridis"
)
````

## 2.4. Verificación de la Calidad de los Datos

Se aplica una auditoría de datos rigurosa para identificar anomalías antes de la fase de preparación:

* **Análisis de Valores Faltantes (*Missings*):** Se cuantifica la presencia de registros nulos en las variables estructuradas. En el componente de texto, se identifican archivos de currículos corruptos, vacíos o que no pudieron ser procesados por el extractor de PDFs.


* **Detección de Valores Atípicos (*Outliers*):** En la variable cuantitativa de puntuación, se inspeccionan valores fuera del rango teórico (0-100) o atípicos univariados mediante diagramas de caja (*Boxplots*) basados en el Rango Intercuartílico ($IQR$). En el texto, se identifican documentos con longitudes anómalas (hojas de vida excesivamente cortas o inusualmente largas) que puedan distorsionar los cálculos de similitud estadística.
* **Consistencia e Integridad:** Se verifica la consistencia referencial: cada `ID` listado en `train.csv` y `test.csv` debe corresponder exactamente con el nombre de un archivo de texto/PDF dentro de sus respectivas carpetas de hojas de vida, garantizando la trazabilidad total del dataset.

````python
def auditar_valores_faltantes(df_train, df_test):
    """Evalúa la presencia de registros nulos en los conjuntos de datos estructurados."""
    print("=" * 80)
    print(" ANALISIS DE VALORES FALTANTES (MISSINGS)")
    print("=" * 80)

    print("\n[1] Valores faltantes en el conjunto de entrenamiento (train.csv):")
    nulos_train = df_train.isnull().sum()
    pct_train = (df_train.isnull().sum() / len(df_train)) * 100
    reporte_train = pd.DataFrame(
        {"Nulos": nulos_train, "Porcentaje (%)": pct_train}
    )
    print(reporte_train[reporte_train["Nulos"] > 0] if nulos_train.sum() > 0 else "No se detectaron valores nulos en train.csv.")

    print("\n[2] Valores faltantes en el conjunto de prueba (test.csv):")
    nulos_test = df_test.isnull().sum()
    pct_test = (df_test.isnull().sum() / len(df_test)) * 100
    reporte_test = pd.DataFrame({"Nulos": nulos_test, "Porcentaje (%)": pct_test})
    print(reporte_test[reporte_test["Nulos"] > 0] if nulos_test.sum() > 0 else "No se detectaron valores nulos en test.csv.")


def auditar_outliers_cuantitativos(df, columna_objetivo):
    """Detecta anomalías y outliers en la variable continua mediante el criterio IQR."""
    print("\n" + "=" * 80)
    print(" DETECCION DE VALORES ATIPICOS (OUTLIERS) - VARIABLE OBJETIVO")
    print("=" * 80)

    if columna_objetivo not in df.columns:
        print(f"La columna '{columna_objetivo}' no se encuentra en el DataFrame.")
        return

    serie = df[columna_objetivo].dropna()

    # Verificación del rango teórico estricto
    fuera_de_rango = df[(df[columna_objetivo] < 0) | (df[columna_objetivo] > 100)]
    print(f"[1] Registros fuera del rango teórico (0-100): {len(fuera_de_rango)}")

    # Cálculo de los límites estadísticos mediante el Rango Intercuartílico
    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    outliers_inf = df[df[columna_objetivo] < limite_inferior]
    outliers_sup = df[df[columna_objetivo] > limite_superior]

    print(f"[2] Criterio Estadístico IQR:")
    print(f"  - Umbral Inferior: {limite_inferior:.4f}%")
    print(f"  - Umbral Superior: {limite_superior:.4f}%")
    print(f"  - Outliers analíticos inferiores detectados: {len(outliers_inf)}")
    print(f"  - Outliers analíticos superiores detectados: {len(outliers_sup)}")


def auditar_corpus_texto(ruta_carpeta, nombre_carpeta):
    """Identifica PDFs corruptos, vacíos o con longitudes anómalas de caracteres."""
    if not os.path.exists(ruta_carpeta):
        print(f"\nRuta no encontrada: {ruta_carpeta}")
        return pd.DataFrame()

    archivos = [f for f in os.listdir(ruta_carpeta) if f.lower().endswith(".pdf")]
    registro_calidad = []

    for archivo in archivos:
        ruta_completa = os.path.join(ruta_carpeta, archivo)
        longitud_caracteres = 0
        estado = "Correcto"

        try:
            reader = PdfReader(ruta_completa)
            texto = ""
            for pagina in reader.pages:
                contenido_pag = pagina.extract_text()
                if contenido_pag:
                    texto += contenido_pag
            longitud_caracteres = len(texto.strip())

            if longitud_caracteres == 0:
                estado = "Vacio / No Extraible"
        except Exception:
            estado = "Corrupto"

        registro_calidad.append(
            {
                "Archivo": archivo,
                "ID_Extraido": archivo.replace(".pdf", ""),
                "Caracteres": longitud_caracteres,
                "Estado": estado,
            }
        )

    df_calidad = pd.DataFrame(registro_calidad)

    print("\n" + "=" * 80)
    print(f" AUDITORIA DE TEXTO: {nombre_carpeta.upper()}")
    print("=" * 80)
    print(f"Total de archivos PDF analizados: {len(df_calidad)}")
    print(f"Archivos corruptos o ilegibles: {len(df_calidad[df_calidad['Estado'] == 'Corrupto'])}")
    print(f"Archivos sin texto extraíble:    {len(df_calidad[df_calidad['Estado'] == 'Vacio / No Extraible'])}")

    # Análisis estadístico de longitudes para identificar anomalías formales
    if len(df_calidad) > 0 and df_calidad["Caracteres"].sum() > 0:
        q1_t = df_calidad["Caracteres"].quantile(0.25)
        q3_t = df_calidad["Caracteres"].quantile(0.75)
        iqr_t = q3_t - q1_t
        lim_inf_t = max(0, q1_t - 1.5 * iqr_t)
        lim_sup_t = q3_t + 1.5 * iqr_t

        anomalos_cortos = df_calidad[df_calidad["Caracteres"] < lim_inf_t]
        anomalos_largos = df_calidad[df_calidad["Caracteres"] > lim_sup_t]

        print(f"Documentos inusualmente cortos (< {lim_inf_t:.1f} carac.): {len(anomalos_cortos)}")
        print(f"Documentos inusualmente largos (> {lim_sup_t:.1f} carac.): {len(anomalos_largos)}")

    return df_calidad


def verificar_consistencia_referencial(df, df_calidad_texto, nombre_set):
    """Valida la integridad referencial exacta entre el archivo estructurado y los PDFs."""
    print("\n" + "=" * 80)
    print(f" CONSISTENCIA REFERENCIAL: {nombre_set.upper()}")
    print("=" * 80)

    # El identificador de la tabla se convierte a string para asegurar homogeneidad
    ids_estructurados = set(df["ID"].astype(str).tolist())
    ids_archivos = set(df_calidad_texto["ID_Extraido"].astype(str).tolist())

    huérfanos_en_tabla = ids_estructurados - ids_archivos
    huérfanos_en_carpetas = ids_archivos - ids_estructurados

    print(f"IDs en la tabla estructurada: {len(ids_estructurados)}")
    print(f"IDs en la carpeta de PDFs:    {len(ids_archivos)}")
    print(f"IDs registrados en tabla pero que NO tienen archivo PDF: {len(huérfanos_en_tabla)}")
    print(f"Archivos PDF en carpeta que NO están registrados en la tabla: {len(huérfanos_en_carpetas)}")


# ==============================================================================
# EJECUCION DE LA AUDITORIA DE CALIDAD
# ==============================================================================

# 1. Ejecución del análisis de nulos
auditar_valores_faltantes(df_train, df_test)

# 2. Ejecución de la detección de outliers en la variable objetivo
auditar_outliers_cuantitativos(df_train, "Match Percentage")

# 3. Auditoría de archivos físicos e integridad de texto
df_calidad_train = auditar_corpus_texto(path_train_resumes, "trainResumes")
df_calidad_test = auditar_corpus_texto(path_test_resumes, "testResumes")

# 4. Verificación de la integridad referencial de los identificadores
if "ID" in df_train.columns and not df_calidad_train.empty:
    verificar_consistencia_referencial(df_train, df_calidad_train, "Train Set")

if "ID" in df_test.columns and not df_calidad_test.empty:
    verificar_consistencia_referencial(df_test, df_calidad_test, "Test Set")
````

# Fase 3: Preparación de los Datos

Esta fase se estructura en las siguientes subtareas metodológicas, las cuales abordan directamente los hallazgos analíticos obtenidos en la auditoría de datos previa:

### 3.1. Limpieza de Datos y Tratamiento de Anomalías

Se conservo los datos en el estado natural por las razones mencionadas previamente.

### 3.2. Preprocesamiento de Texto (Normalización)

Antes de proceder con la transformación matemática, el texto plano extraído de los 150 archivos PDF (90 de entrenamiento y 60 de prueba) se somete a un pipeline secuencial de normalización:

* **Conversión a minúsculas:** Homogeneización de todas las cadenas de caracteres para eliminar la distinción sintáctica entre grafías idénticas (ej. transformar "Machine", "machine" y "MACHINE" a una única representación estándar).
* **Remoción de caracteres especiales y puntuación:** Depuración de guiones, viñetas, símbolos derivados de la decodificación del PDF y caracteres numéricos aislados que carecen de valor semántico para el perfil.
* **Eliminación de palabras vacías (*Stop-Words*):** Filtrado y exclusión de conectores gramaticales, artículos y preposiciones en los idiomas inglés y español mediante el uso de librerías especializadas como `NLTK` o `SpaCy`.
* **Lematización:** Reducción de las palabras flexionadas a su raíz morfológica o lema básico, logrando que términos equivalentes (ej. "desarrollando", "desarrolló" y "desarrolla") se unifiquen bajo una misma entidad conceptual ("desarrollar").


````python
# ==============================================================================
# CONFIGURACIÓN DE RECURSOS LINGÜÍSTICOS (SpaCy)
# ==============================================================================

# Carga de modelos para análisis lingüístico y lematización
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# Extracción de stop-words nativas de ambos modelos
stopwords_global = nlp_es.Defaults.stop_words.union(nlp_en.Defaults.stop_words)

# Incorporación de términos corporativos irrelevantes identificados en el EDA
terminos_adicionales = {
    "correo",
    "teléfono",
    "telefono",
    "experiencia",
    "año",
    "años",
    "perfil",
    "contacto",
    "habilidades",
    "idiomas",
    "educación",
    "currículum",
}
stopwords_global = stopwords_global.union(terminos_adicionales)


# ==============================================================================
# PIPELINE SECUENCIAL DE NORMALIZACIÓN DE TEXTO
# ==============================================================================


def normalizar_y_lematizar_texto(texto_sucio):
    """Ejecuta la limpieza, remoción de caracteres no semánticos y reducción

    morfológica mediante lematización condicionada por idioma.
    """
    if not texto_sucio or len(texto_sucio.strip()) == 0:
        return ""

    # 1. Conversión a minúsculas
    texto = texto_sucio.lower()

    # 2. Remoción de ruido estructural (URLs, correos y caracteres especiales)
    texto = re.sub(r"https?://\S+|www\.\S+", "", texto)
    texto = re.sub(r"\S+@\S+", "", texto)

    # Conservación exclusiva de caracteres alfabéticos de los alfabetos español e inglés
    texto = re.sub(r"[^a-záéíóúüñ\s]", " ", texto)
    texto = re.sub(r"\s+", " ", texto).strip()

    # 3. Procesamiento lingüístico dinámico (Detección simple de idioma)
    palabras_clave_es = {"y", "o", "en", "con", "para", "del", "por", "sobre"}
    tokens_iniciales = set(texto.split()[:20])

    if tokens_iniciales.intersection(palabras_clave_es):
        doc = nlp_es(texto)
    else:
        doc = nlp_en(texto)

    # 4. Extracción de lemas y filtrado selectivo de Stop-Words
    tokens_limpios = [
        token.lemma_
        for token in doc
        if token.lemma_ not in stopwords_global and len(token.lemma_) > 2
    ]

    return " ".join(tokens_limpios)


def extraer_texto_crudo_pdf(ruta_pdf):
    """Extrae las cadenas de caracteres almacenadas en las páginas del PDF."""
    try:
        reader = PdfReader(ruta_pdf)
        texto_acumulado = ""
        for pagina in reader.pages:
            contenido = pagina.extract_text()
            if contenido:
                texto_acumulado += contenido + " "
        return texto_acumulado.strip()
    except Exception:
        return ""


def procesar_repositorio_hojas_vida(df_estructurado, ruta_carpeta):
    """Asocia de forma biunívoca cada registro de la tabla con su PDF correspondiente,

    utilizando la columna rectificada 'CandidateID' para la trazabilidad.
    """
    if "CandidateID" not in df_estructurado.columns:
        raise ValueError(
            "El DataFrame provisto no contiene la columna rectificada 'CandidateID'."
        )

    textos_normalizados = []
    total_registros = len(df_estructurado)

    print(
        f"\nIniciando el procesamiento de {total_registros} documentos en: {os.path.basename(ruta_carpeta)}"
    )

    for idx, row in df_estructurado.iterrows():
        # Rectificación: Uso explícito de 'CandidateID' en lugar de 'ID'
        id_candidato = str(row["CandidateID"])
        ruta_pdf = os.path.join(ruta_carpeta, f"{id_candidato}.pdf")

        if os.path.exists(ruta_pdf):
            texto_crudo = extraer_texto_crudo_pdf(ruta_pdf)
            texto_limpio = normalizar_y_lematizar_texto(texto_crudo)
            textos_normalizados.append(texto_limpio)
        else:
            textos_normalizados.append("")

        # Indicador visual de progreso
        if (idx + 1) % 30 == 0 or (idx + 1) == total_registros:
            print(f"  -> Procesados: {idx + 1} de {total_registros} documentos.")

    return textos_normalizados


# ==============================================================================
# EJECUCIÓN SÍNCRONA DE LA FASE 3 CON COLUMNAS RECTIFICADAS
# ==============================================================================

# --- Subtarea 3.1: Tratamiento de Anomalías Estructuradas ---
# Conservación explícita de la variable objetivo continua 'Match Percentage'
df_train_preparado = df_train.copy()
df_test_preparado = df_test.copy()
print(
    "Subtarea 3.1: Tratamiento finalizado. Registros cuantitativos conservados en escala original."
)

# --- Subtarea 3.2: Ejecución del Pipeline Textual ---
# Procesamiento del conjunto de entrenamiento utilizando la columna rectificada
df_train_preparado["Texto_Limpio"] = procesar_repositorio_hojas_vida(
    df_train_preparado, path_train_resumes
)

# Procesamiento del conjunto de prueba utilizando la columna rectificada
df_test_preparado["Texto_Limpio"] = procesar_repositorio_hojas_vida(
    df_test_preparado, path_test_resumes
)

# Procesamiento individual del Perfil del Puesto (Job Description)
if os.path.exists(path_pdf):
    texto_job_crudo = extraer_texto_crudo_pdf(path_pdf)
    perfil_puesto_limpio = normalizar_y_lematizar_texto(texto_job_crudo)
    print(
        "\nSubtarea 3.2: Procesamiento del archivo 'Job description.pdf' finalizado con éxito."
    )
else:
    perfil_puesto_limpio = ""
    print("\nAdvertencia: El archivo del perfil del puesto no fue localizado.")

# --- Comprobación de la Integración final de los datos ---
print("\n" + "=" * 80)
print(" VERIFICACIÓN DE LAS MATRICES PREPARADAS")
print("=" * 80)
print(f"Estructura final de Entrenamiento: {df_train_preparado.shape}")
print(f"Estructura final de Prueba:        {df_test_preparado.shape}")
print("\nMuestra de texto normalizado del primer candidato (Entrenamiento):")
print(df_train_preparado["Texto_Limpio"].iloc[0][:300] + "...")
````


### 3.3. Ingeniería de Características (*Feature Engineering* / Vectorización)

Esta subtarea constituye el núcleo analítico de la Fase 3, donde el texto normalizado se transforma en estructuras numéricas aptas para el procesamiento algorítmico:

* **Pipeline de Extracción y Representación Espacial:** Transformación de las colecciones de texto en matrices cuantitativas. Se evalúa el uso de esquemas de pesaje estadístico como **TF-IDF** (*Term Frequency - Inverse Document Frequency*) o, en su defecto, la generación de **Word/Sentence Embeddings** densos basados en modelos de lenguaje preentrenados para capturar de manera rigurosa las relaciones semánticas del entorno técnico.
* **Alineación con la Descripción del Puesto:** El documento maestro `Job description.pdf` se procesa exactamente a través del mismo pipeline de normalización y vectorización diseñado para los currículos. Esto permite proyectarlo dentro del mismo espacio vectorial y transformarlo en el **vector de referencia ideal** que servirá de contraste en las fases posteriores.


````python
def ejecutar_ingenieria_caracteristicas(
    df_train, df_test, texto_perfil_puesto
):
    """Transforma el texto normalizado en representaciones vectoriales densas

    mediante el esquema de pesaje estadístico TF-IDF.
    """
    print("=" * 80)
    print(" SUBTAREA 3.3: INGENIERÍA DE CARACTERÍSTICAS / VECTORIZACIÓN")
    print("=" * 80)

    # 1. Validación de la existencia de datos procesados en la fase previa
    if (
        "Texto_Limpio" not in df_train.columns
        or "Texto_Limpio" not in df_test.columns
    ):
        raise ValueError(
            "No se localizó la columna 'Texto_Limpio' en las matrices de datos."
        )

    if not texto_perfil_puesto:
        raise ValueError(
            "El texto normalizado del perfil del puesto se encuentra vacío."
        )

    # 2. Configuración del Vectorizador TF-IDF
    # Se seleccionan unigramas y bigramas (1, 2) para capturar términos compuestos (ej. 'machine learning')
    # Se limita a las 1500 características más significativas para evitar la maldición de la dimensionalidad
    tfidf_vectorizer = TfidfVectorizer(
        ngram_range=(1, 2), max_features=1500, min_df=2
    )

    print("\n[1] Inicialización del modelo TF-IDF con rango de (1, 2) n-gramas.")

    # 3. Ajuste y transformación sobre el conjunto de entrenamiento
    # El vocabulario se aprende exclusivamente del set de entrenamiento para evitar el sesgo de información
    X_train_tfidf = tfidf_vectorizer.fit_transform(df_train["Texto_Limpio"])

    # 4. Transformación del conjunto de prueba bajo el mismo espacio vectorial
    X_test_tfidf = tfidf_vectorizer.transform(df_test["Texto_Limpio"])

    # 5. Alineación y proyección del vector de referencia ideal (Job Description)
    # Se aplica el método transform para garantizar que use exactamente las mismas dimensiones y términos
    vector_perfil_ideal = tfidf_vectorizer.transform([texto_perfil_puesto])

    print("[2] Proceso de vectorización y alineación espacial completado.")

    # 6. Diagnóstico cuantitativo de las matrices generadas
    print(
        f"  - Dimensiones de la matriz de entrenamiento: {X_train_tfidf.shape}"
    )
    print(f"  - Dimensiones de la matriz de prueba:        {X_test_tfidf.shape}")
    print(
        f"  - Dimensiones del vector de referencia ideal: {vector_perfil_ideal.shape}"
    )

    # 7. Extracción de términos de alta relevancia para validación analítica
    caracteristicas = tfidf_vectorizer.get_feature_names_out()
    indices_altos = np.argsort(vector_perfil_ideal.toarray()[0])[::-1][:10]

    print("\n[3] Variables de control: Top 10 tokens con mayor peso en el perfil ideal:")
    for rango, idx in enumerate(indices_altos, start=1):
        peso = vector_perfil_ideal.toarray()[0][idx]
        print(f"  {rango}. Término: '{caracteristicas[idx]}' | Peso TF-IDF: {peso:.4f}")

    return X_train_tfidf, X_test_tfidf, vector_perfil_ideal, tfidf_vectorizer


# ==============================================================================
# EJECUCIÓN SÍNCRONA DE LA SUBTAREA 3.3
# ==============================================================================

X_train_vectorizado, X_test_vectorizado, v_perfil_ideal, vectorizador_fitted = (
    ejecutar_ingenieria_caracteristicas(
        df_train_preparado, df_test_preparado, perfil_puesto_limpio
    )
)
````


### 3.4. Integración y Formateo Final

* **Consolidación Matricial:** Unión de la matriz numérica resultante de la vectorización del texto con los vectores de identificadores (`ID`) y los valores continuos de la variable objetivo provenientes de `train.csv`.
* **Aseguramiento de la Consistencia Dimensional:** Verificación de que el espacio vectorial y las dimensiones de las columnas sean estrictamente idénticos tanto para el conjunto de entrenamiento como para el de prueba (`test.csv`). Esto garantiza que el vocabulario indexado sea homogéneo y previene discrepancias estructurales al momento de alimentar los modelos en la fase de modelado.

````python
def ejecutar_integracion_y_formateo_final(
    df_train, df_test, X_train_tfidf, X_test_tfidf
):
    """Consolida las características vectoriales, aísla los identificadores

    y verifica la homogeneidad dimensional antes de la fase de modelado.
    """
    print("=" * 80)
    print(" SUBTAREA 3.4: INTEGRACIÓN Y FORMATEO FINAL")
    print("=" * 80)

    # 1. Aislamiento de los identificadores como vectores independientes de control
    # Se evita la inserción en la matriz dispersa para prevenir conflictos de tipos de datos
    ids_train = df_train["CandidateID"].astype(str).values
    ids_test = df_test["CandidateID"].astype(str).values

    # 2. Extracción de la variable objetivo continua (vector de etiquetas)
    y_train = df_train["Match Percentage"].values

    print("\n[1] Control de Calidad Dimensional y Homogeneidad:")

    # 3. Verificación de que el espacio de características TF-IDF coincida en ambos conjuntos
    dimension_train = X_train_tfidf.shape[1]
    dimension_test = X_test_tfidf.shape[1]

    if dimension_train == dimension_test:
        print(
            f"  - Validación: Espacio vectorial homogéneo ({dimension_train} características numéricas)."
        )
    else:
        raise ValueError(
            f"Discrepancia detectada: Entrenamiento posee {dimension_train} columnas y Prueba posee {dimension_test}."
        )

    print("\n[2] Resumen de Estructuras Finales Listas para Modelamiento:")
    print(
        f"  - Matriz de entrenamiento (X_train) ➔ Observaciones: {X_train_tfidf.shape[0]} | Características: {X_train_tfidf.shape[1]}"
    )
    print(
        f"  - Vector objetivo (y_train)         ➔ Registro de etiquetas continuas: {y_train.shape[0]}"
    )
    print(
        f"  - Matriz de prueba (X_test)        ➔ Observaciones: {X_test_tfidf.shape[0]} | Características: {X_test_tfidf.shape[1]}"
    )
    print(
        f"  - Vectores de control referencial  ➔ IDs Entrenamiento: {ids_train.shape[0]} | IDs Prueba: {ids_test.shape[0]}"
    )

    return X_train_tfidf, y_train, X_test_tfidf, ids_train, ids_test


# ==============================================================================
# EJECUCIÓN SÍNCRONA DE LA SUBTAREA 3.4 RECTIFICADA
# ==============================================================================

X_train_final, y_train_final, X_test_final, ids_train_control, ids_test_control = (
    ejecutar_integracion_y_formateo_final(
        df_train_preparado,
        df_test_preparado,
        X_train_vectorizado,
        X_test_vectorizado,
    )
)
````

# Fase 4: Modelización (Modeling)

Bajo el rigor de la metodología CRISP-DM y considerando la naturaleza de tu proyecto, se debe diseñar una estrategia adaptada a un problema de **regresión predictiva** con datos vectoriales de texto de alta dimensionalidad (1454 características).

Dado que el objetivo es predecir una variable continua (`Match Percentage`), la estrategia debe enfocarse en algoritmos de regresión y en métricas de error de escala lineal, evitando métricas de clasificación (como matrices de confusión o F1-Score) que no aplican a variables continuas.

A continuación, se detallan las subtareas metodológicas y las directrices técnicas que se deben implementar en el código para esta fase:


## 4.1. Selección de Técnicas de Modelado

Dada la estructura matricial dispersa y el tamaño muestral ($N = 90$ en entrenamiento), se deben seleccionar algoritmos con diferentes enfoques matemáticos para evaluar su capacidad de generalización y robustez ante el riesgo de sobreajuste (*overfitting*):

1. **Línea Base Lineal Penalizada (Ridge o Lasso Regression):** Modelo lineal con regularización (L1).
2. **Algoritmos Basados en Árboles de Decisión (Random Forest Regressor):** Permite capturar interacciones no lineales entre los términos (*tokens*) del texto sin necesidad de asumir una distribución normal en los datos.
3. **Ensamblado de Gradiente Avanzado (XGBoost):** Modelo secuencial optimizados para minimizar el error de forma iterativa, ideales para manejar la bimodalidad identificada en la exploración de datos.



## 4.2. Diseño del Esquema de Prueba (Validación)

Para mitigar el riesgo de sobreajuste (*overfitting*) derivado de la alta dimensionalidad ($1454$ columnas) en contraste con el tamaño muestral acotado del conjunto de entrenamiento ($N = 90$), no es metodológicamente viable aplicar una partición simple (*Hold-Out Validation*). En su lugar, se implementa una estrategia de **Validación Cruzada de K-Pliegues Repetida (*Repeated K-Fold Cross-Validation*)**:

* **Configuración del Esquema:** Se establece un diseño de $K = 5$ pliegues con $3$ repeticiones, lo que genera un total de $15$ iteraciones independientes de entrenamiento y validación. Esto asegura que cada observación sea evaluada en múltiples subconjuntos, estabilizando la varianza de las métricas de rendimiento estimadas.
* **Prevención de Filtración de Información (*Data Leakage*):** Dado que la matriz de características `X_train_final` ya fue vectorizada de manera pasiva bajo el vocabulario estricto del set de entrenamiento en la Fase 3, el entorno se encuentra blindado contra la filtración de información entre los pliegues de validación y entrenamiento.

## 4.3. Construcción y Optimización de Modelos

El proceso de entrenamiento de los tres algoritmos seleccionados se ejecuta mediante búsquedas iterativas para localizar las combinaciones óptimas de hiperparámetros, evitando el uso de configuraciones por defecto que puedan inducir a un sesgo de subajuste o sobreajuste:

* **Sintonización de Hiperparámetros (Grid Search / Random Search):**
* **Lasso Regression (L1):** Se optimiza el parámetro de regularización alpha ($\alpha$), el cual controla la fuerza de la penalización encargada de colapsar a cero los coeficientes de los tokens de texto irrelevantes, actuando como un selector de características embebido.
* **Random Forest Regressor:** Se evalúa el número óptimo de estimadores (`n_estimators`) y la profundidad máxima de los árboles (`max_depth`) para controlar la varianza del ensamble por embolsamiento (*bagging*).
* **XGBoost Regressor:** Se sintoniza la tasa de aprendizaje (`learning_rate`), la fracción de submuestreo de columnas (`colsample_bytree`) y la profundidad de los árboles, maximizando la capacidad del algoritmo para corregir los residuos de forma secuencial sobre la distribución bimodal.


## 4.4. Evaluación Técnica del Modelo

La validación del rendimiento de los regresores se fundamenta estrictamente en la teoría estadística de variables continuas, midiendo la magnitud física de las desviaciones con respecto al porcentaje real de adaptabilidad (`Match Percentage` en la escala $0-100\%$):

* **Error Absoluto Medio (MAE - *Mean Absolute Error*):** Cuantifica la magnitud promedio del error directamente en las unidades del negocio (puntos porcentuales de adaptabilidad). Esta métrica permite una interpretación intuitiva de qué tan distante se encuentra la predicción del valor real en promedio.
* **Raíz del Error Cuadrático Medio (RMSE - *Root Mean Squared Error*):** Al incorporar un factor cuadrático, penaliza de manera severa las desviaciones de gran magnitud. Es la métrica crítica de control para el negocio, puesto que el sistema debe minimizar los errores graves de sobreestimación o subestimación en el perfil del candidato.
* **Coeficiente de Determinación ($R^2$):** Mide la proporción de la varianza de la variable objetivo que es explicada de forma predictiva por las $1454$ variables de texto normalizadas y lematizadas.


````python
# ==============================================================================
# CONFIGURACIÓN DEL ESQUEMA DE VALIDACIÓN (SUBTAREA 4.2)
# ==============================================================================

# Se define un esquema de Validación Cruzada de K-Pliegues Repetida
# 5 pliegues con 3 repeticiones genera 15 iteraciones de evaluación independiente.
# Se fija una semilla aleatoria (random_state) para asegurar la reproducibilidad.
semilla = 42
rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=semilla)

print("=" * 80)
# Se evita el uso de caracteres informales en la salida de consola
print(" INICIANDO FASE 4: OPTIMIZACIÓN Y ENTRENAMIENTO DE MODELOS")
print("=" * 80)
print(f"Esquema de validación configurado: 5-Fold Cross-Validation Repetido (3 veces). Total iteraciones: 15.")

# Dicionario global para almacenar los resultados analíticos de cada algoritmo
reporte_tecnico = {}

# ==============================================================================
# 1. OPTIMIZACIÓN Y CONSTRUCCIÓN DE LASSO REGRESSION (SUBTAREA 4.1 y 4.3)
# ==============================================================================
print("\n" + "-" * 50)
print(" Ejecutando optimización para: Lasso Regression (L1)")
print("-" * 50)

# El parámetro 'alpha' controla la fuerza de la penalización L1
param_grid_lasso = {
    'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]
}

# Se utiliza GridSearchCV sobre las 15 particiones de la validación cruzada
grid_lasso = GridSearchCV(
    estimator=Lasso(random_state=semilla, max_iter=2000),
    param_grid=param_grid_lasso,
    cv=rkf,
    scoring='neg_mean_absolute_error',
    n_jobs=-1
)

# Ajuste del modelo sobre la matriz de entrenamiento de la Fase 3
grid_lasso.fit(X_train_final, y_train_final)

# Extracción de las métricas obtenidas en la mejor configuración
mejor_lasso = grid_lasso.best_estimator_
mae_cv_lasso = abs(grid_lasso.best_score_)

# Cálculo de predicciones fuera de muestra (entrenamiento) para computar RMSE y R2 promedio aproximado
y_pred_train_lasso = mejor_lasso.predict(X_train_final)
rmse_lasso = np.sqrt(mean_squared_error(y_train_final, y_pred_train_lasso))
r2_lasso = r2_score(y_train_final, y_pred_train_lasso)

reporte_tecnico['Lasso Regression'] = {
    'Hiperparámetros Óptimos': grid_lasso.best_params_,
    'MAE (Validación Cruzada)': mae_cv_lasso,
    'RMSE (Entrenamiento)': rmse_lasso,
    'R² (Entrenamiento)': r2_lasso,
    'Objeto_Modelo': mejor_lasso
}

print(f"  -> Configuración óptima seleccionada: {grid_lasso.best_params_}")
print(f"  -> MAE promedio obtenido en validación cruzada: {mae_cv_lasso:.4f}%")

# ==============================================================================
# 2. OPTIMIZACIÓN Y CONSTRUCCIÓN DE RANDOM FOREST REGRESSOR (SUBTAREA 4.1 y 4.3)
# ==============================================================================
print("\n" + "-" * 50)
print(" Ejecutando optimización para: Random Forest Regressor")
print("-" * 50)

# Espacio de búsqueda para el algoritmo de ensamble por embolsamiento
param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

grid_rf = GridSearchCV(
    estimator=RandomForestRegressor(random_state=semilla),
    param_grid=param_grid_rf,
    cv=rkf,
    scoring='neg_mean_absolute_error',
    n_jobs=-1
)

grid_rf.fit(X_train_final, y_train_final)

mejor_rf = grid_rf.best_estimator_
mae_cv_rf = abs(grid_rf.best_score_)

y_pred_train_rf = mejor_rf.predict(X_train_final)
rmse_rf = np.sqrt(mean_squared_error(y_train_final, y_pred_train_rf))
r2_rf = r2_score(y_train_final, y_pred_train_rf)

reporte_tecnico['Random Forest'] = {
    'Hiperparámetros Óptimos': grid_rf.best_params_,
    'MAE (Validación Cruzada)': mae_cv_rf,
    'RMSE (Entrenamiento)': rmse_rf,
    'R² (Entrenamiento)': r2_rf,
    'Objeto_Modelo': mejor_rf
}

print(f"  -> Configuración óptima seleccionada: {grid_rf.best_params_}")
print(f"  -> MAE promedio obtenido en validación cruzada: {mae_cv_rf:.4f}%")

# ==============================================================================
# 3. OPTIMIZACIÓN Y CONSTRUCCIÓN DE XGBOOST REGRESSOR (SUBTAREA 4.1 y 4.3)
# ==============================================================================
print("\n" + "-" * 50)
print(" Ejecutando optimización para: XGBoost Regressor")
print("-" * 50)

# Espacio de búsqueda para el algoritmo de aumento de gradiente secuencial
param_grid_xgb = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'colsample_bytree': [0.7, 0.9]
}

grid_xgb = GridSearchCV(
    estimator=xgb.XGBRegressor(objective='reg:squarederror', random_state=semilla),
    param_grid=param_grid_xgb,
    cv=rkf,
    scoring='neg_mean_absolute_error',
    n_jobs=-1
)

grid_xgb.fit(X_train_final, y_train_final)

mejor_xgb = grid_xgb.best_estimator_
mae_cv_xgb = abs(grid_xgb.best_score_)

y_pred_train_xgb = mejor_xgb.predict(X_train_final)
rmse_xgb = np.sqrt(mean_squared_error(y_train_final, y_pred_train_xgb))
r2_xgb = r2_score(y_train_final, y_pred_train_xgb)

reporte_tecnico['XGBoost'] = {
    'Hiperparámetros Óptimos': grid_xgb.best_params_,
    'MAE (Validación Cruzada)': mae_cv_xgb,
    'RMSE (Entrenamiento)': rmse_xgb,
    'R² (Entrenamiento)': r2_xgb,
    'Objeto_Modelo': mejor_xgb
}

print(f"  -> Configuración óptima seleccionada: {grid_xgb.best_params_}")
print(f"  -> MAE promedio obtenido en validación cruzada: {mae_cv_xgb:.4f}%")

# ==============================================================================
# REPORTE DE EVALUACIÓN TÉCNICA FINAL
# ==============================================================================
print("\n" + "=" * 80)
print(" REPORTE DE EVALUACIÓN TÉCNICA FINAL (CONSOLIDADO)")
print("=" * 80)

# Conversión del diccionario de resultados a un DataFrame estructurado
datos_reporte = []
for modelo_nombre, metricas in reporte_tecnico.items():
    # Se corrigen los cierres de paréntesis internos de las claves del diccionario
    datos_reporte.append(
        {
            "Modelo Algorítmico": modelo_nombre,
            "Métrica MAE (CV)": f"{metricas['MAE (Validación Cruzada)']:.4f}%",
            "Métrica RMSE (Train)": f"{metricas['RMSE (Entrenamiento)']:.4f}%",
            "Coeficiente R² (Train)": f"{metricas['R² (Entrenamiento)']:.4f}",
            "Hiperparámetros Seleccionados": str(
                metricas["Hiperparámetros Óptimos"]
            ),
        }
    )

df_comparativo_modelos = pd.DataFrame(datos_reporte)
# Despliegue de la tabla comparativa limpia en la libreta
print(df_comparativo_modelos.to_string(index=False))

# Identificación automática del modelo campeón basado en el menor MAE de validación cruzada
nombre_campeon = min(
    reporte_tecnico,
    key=lambda k: reporte_tecnico[k]["MAE (Validación Cruzada)"],
)
modelo_campeon_objeto = reporte_tecnico[nombre_campeon]["Objeto_Modelo"]

print("\n" + "=" * 80)
print(f" SELECCIÓN DE MODELO CAMPEÓN: {nombre_campeon.upper()}")
print("=" * 80)
print(
    f"El algoritmo seleccionado para la ejecución de inferencia en la fase de evaluación es '{nombre_campeon}', "
)
print(
    "debido a que registró el menor error absoluto medio (MAE) general durante las pruebas cruzadas."
)
````

Prediccón de los valores para el conjunto de testeo:

````python
def ejecutar_inferencia_testeo(modelo_campeon, X_test_sparse, ids_test_control):
    """Genera las predicciones del Match Percentage sobre el conjunto de prueba

    y consolida el archivo final de resultados asegurando el rango de negocio.
    """
    print("=" * 80)
    print(" FASE 5: EJECUCIÓN DE INFERENCIA SOBRE EL CONJUNTO DE TESTEO")
    print("=" * 80)

    # 1. Ejecución de la predicción algorítmica fuera de muestra
    predicciones_raw = modelo_campeon.predict(X_test_sparse)

    # 2. Post-procesamiento y acotamiento al rango teórico del negocio (0-100%)
    # Esto previene que eventuales extrapolaciones del modelo lineal generen valores inválidos
    predicciones_acotadas = [max(0.0, min(100.0, p)) for p in predicciones_raw]

    # 3. Consolidación de la estructura estructurada para la entrega
    df_resultados_test = pd.DataFrame(
        {
            "CandidateID": ids_test_control,
            "Match Percentage": predicciones_acotadas,
        }
    )

    print("\n[1] Control de Calidad de la Inferencia Realizada:")
    print(f"  - Total de candidatos evaluados: {len(df_resultados_test)}")
    print(
        f"  - Valor mínimo asignado:        {df_resultados_test['Match Percentage'].min():.4f}%"
    )
    print(
        f"  - Valor máximo asignado:        {df_resultados_test['Match Percentage'].max():.4f}%"
    )
    print(
        f"  - Promedio general estimado:    {df_resultados_test['Match Percentage'].mean():.4f}%"
    )

    # 4. Muestra analítica para verificación de consistencia de los datos
    print("\n[2] Primeros 10 registros del conjunto de prueba completados:")
    print(df_resultados_test.head(10).to_string(index=False))

    return df_resultados_test


# ==============================================================================
# EJECUCIÓN SÍNCRONA DE LA INFERENCIA CON EL MODELO CAMPEÓN
# ==============================================================================

# Se invoca la función pasando el objeto del modelo guardado automáticamente en la fase 4
df_final_entregable = ejecutar_inferencia_testeo(
    modelo_campeon=modelo_campeon_objeto,
    X_test_sparse=X_test_final,
    ids_test_control=ids_test_control,
)

# 5. Exportación del archivo CSV para la fase de implementación final
df_final_entregable.to_csv("predictions_test.csv", index=False)
````

* **Opción B: Despliegue mediante Vercel (Integración Continua)**

2. En el panel de control, se presiona la opción para añadir un nuevo proyecto (*Add New* ➔ *Project*).

4. En la sección de configuración del marco de trabajo (*Framework Preset*), se selecciona la opción *Other* o *Static HTML*.
5. Se presiona el botón de despliegue (*Deploy*). La plataforma compilará el sitio en un servidor de distribución global y asignará un dominio público seguro de inmediato.
